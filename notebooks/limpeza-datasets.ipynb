```python
import pandas as pd
```

```python
mosquito = pd.read_csv('Mosquito.Borne.Disease.csv')
```

```python
mosquito.head()
```

```python
dados_uteis = mosquito[['NM_BAIRRO', 'NM_DISEASE', 'DT_NOTIFIC']]
```

```python
dados_uteis.head()
```

```python
dados_uteis = dados_uteis.dropna()
```

```python
dados_uteis.columns = ['bairro', 'doenca', 'dia']
```

```python
dados_uteis
```

```python
populacao_por_bairro = pd.read_csv('populacao-por-bairro.csv')
```

```python
from unidecode import unidecode
```

```python
populacao_por_bairro['bairro'] = populacao_por_bairro['bairro'].apply(unidecode)
```

```python
populacao_por_bairro['bairro']
```

```python
populacao_por_bairro['bairro'] = populacao_por_bairro['bairro'].str.upper()
```

```python
casos_por_bairro = dados_uteis.merge(populacao_por_bairro, how='inner', on='bairro')
```
